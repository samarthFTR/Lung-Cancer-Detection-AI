# Changing Data types 

```python
categorical_cols = [
    'Gender', 'Genetic Risk', 'Balanced Diet', 'Obesity', 'Smoking',
    'Passive Smoker', 'Chest Pain', 'Coughing of Blood', 'Fatigue',
    'Weight Loss', 'Shortness of Breath', 'Wheezing', 'Swallowing Difficulty',
    'Clubbing of Finger Nails', 'Frequent Cold', 'Dry Cough', 'Snoring', 'Level',
    'Air Pollution', 'Alcohol use', 'Dust Allergy', 'OccuPational Hazards',
    'chronic Lung Disease'
]

df[categorical_cols] = df[categorical_cols].astype('category')
```

```python
df['Gender']=df['Gender'].map({1:'Male',2:'Female'})
```

```python
df.to_csv('dataset\\cancer patient data sets.csv')
```

```python
df.drop("Patient Id",axis=1,inplace=True)
df_1=df.drop("Level",axis=1,inplace=False)
```

```python
df_1.columns
```

```python
df_1.drop('Gender',axis=1,inplace=True)
sns.heatmap(df_1.corr(method="pearson"))
plt.show()
```

```python
counts = df['Level'].value_counts()
plt.title("Distribution of High,Medium and Low Risk in dataset")
plt.pie(counts, labels=counts.index, autopct='%1.1f%%',shadow=True)
plt.show()
```
