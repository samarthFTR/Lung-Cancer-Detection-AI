## Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dataset\\cancer patient data sets.csv")
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.drop("Patient Id",axis=1,inplace=True)
df_1=df.drop("Level",axis=1,inplace=False)
```

```python
df_1.columns
```

```python
sns.heatmap(df_1.corr(method="pearson"))
plt.show()
```

```python
counts = df['Level'].value_counts()
plt.title("Distribution of High,Medium and Low Risk in dataset")
plt.pie(counts, labels=counts.index, autopct='%1.1f%%',shadow=True)
plt.show()
```
